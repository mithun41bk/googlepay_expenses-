```python
class ExpenseSharing:
    def __init__(self, friends):
        self.friends = friends
        # Initialize everyone's balance at 0
        self.balances = {friend: 0 for friend in friends}

    def add_expense(self, payer, amount, participants):
        split_amount = amount / len(participants)

        # Subtract the share from each participant
        for participant in participants:
            if participant in self.balances:
                self.balances[participant] -= split_amount

        # Add the total amount back to the payer's balance
        if payer in self.balances:
            self.balances[payer] += amount

    def calculate_settlements(self):
        creditors = []
        debtors = []

        # Separate people into those who owe and those who are owed
        for friend, balance in self.balances.items():
            if balance > 0.01: # Use a small threshold for float precision
                creditors.append([friend, balance])
            elif balance < -0.01:
                debtors.append([friend, abs(balance)])

        print(f"\n--- Final Balances ---")
        for friend, bal in self.balances.items():
            print(f"{friend}: {'₹' if bal >= 0 else '-₹'}{abs(bal):.2f}")

        print("\n--- Optimized Settlements ---")

        # Match debtors with creditors
        while debtors and creditors:
            debtor_info = debtors[0]
            creditor_info = creditors[0]

            payment = min(debtor_info[1], creditor_info[1])

            print(f"**{debtor_info[0]} owes {creditor_info[0]}: ₹{payment:.2f}**")

            # Update remaining amounts
            debtor_info[1] -= payment
            creditor_info[1] -= payment

            # Remove from list if settled
            if debtor_info[1] < 0.01:
                debtors.pop(0)
            if creditor_info[1] < 0.01:
                creditors.pop(0)

# --- Driver Code ---
if __name__ == "__main__":
    raw_friends = input("Enter the names of friends, separated by commas: ")
    friends_list = [f.strip() for f in raw_friends.split(",")]

    expense_sharing = ExpenseSharing(friends_list)

    while True:
        payer = input("\nWho paid? (or type 'done' to calculate): ").strip()
        if payer.lower() == "done":
            break

        if payer not in friends_list:
            print("Error: Name not in the original friends list.")
            continue

        try:
            amount = float(input(f"How much did {payer} pay? ₹"))

            raw_parts = input("Who shared this expense? (comma separated): ")
            participants = [p.strip() for p in raw_parts.split(",")]

            expense_sharing.add_expense(payer, amount, participants)
        except ValueError:
            print("Invalid input. Please enter a numerical amount.")

    expense_sharing.calculate_settlements()
```
